## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
def calculate_subtotal(row):
    return row["unit_price"] * row["qty"]

df["subtotal"] = df.apply(calculate_subtotal, axis=1)
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

def calculate_shipping_price(row):
    if row["unit_weight"] * row["qty"] > 50:
        return row["unit_weight"] * row["qty"] * 7
    else:
        return row["unit_weight"] * row["qty"] * 10

df["shipping_price"] = df.apply(calculate_shipping_price, axis=1)
df.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
pre_tax_total = df["subtotal"] + df["shipping_price"]
sales_tax = pre_tax_total * .0925
df["total_price"] = round(pre_tax_total + sales_tax, 2)

df.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df["cost"] = df["unit_cost"] * df["qty"] + df["shipping_price"]
df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
df["profit"] = df["total_price"] - df["cost"]
df.head(20)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
"""
33615    220
66037    211
46820    209
38378    207
24741    207
"""

client_ids = [33615, 66037, 46820, 38378, 24741]

for client_id in client_ids:
    filtered_df = df[df["client_id"] == client_id]
    total_price = filtered_df["total_price"].sum()
    print(f"Client {client_id} had a total price of ${total_price:.2f}")

```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
client_ids = [33615, 66037, 46820, 38378, 24741]

# Filter the DataFrame for the specified client_ids
filtered_df = df[df["client_id"].isin(client_ids)]

# Calculate the totals for each client
summary_df = filtered_df.groupby("client_id").agg({
    "qty": "sum",
    "shipping_price": "sum",
    "total_price": "sum",
    "profit": "sum"
}).reset_index()

# Sort the summary DataFrame by total profit
summary_df = summary_df.sort_values("profit", ascending=False)

summary_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
summary_df["Units"] = summary_df["qty"]
summary_df["Shipping"] = summary_df["shipping_price"].apply(lambda x: "${:.2f}M".format(x / 1000000))
summary_df["Total Revenue"] = summary_df["total_price"].apply(lambda x: "${:.2f}M".format(x / 1000000))
summary_df["Total Profit"] = summary_df["profit"].apply(lambda x: "${:.2f}M".format(x / 1000000))

summary_df = summary_df.rename(columns={"client_id": "Client ID", "Units": "Units", "Shipping": "Shipping", "Total Revenue": "Total Revenue", "Total Profit": "Total Profit"})

summary_df = summary_df[["Client ID", "Units", "Shipping", "Total Revenue", "Total Profit"]]

summary_df

```

```python
# Sort the updated data by "Total Profit" form highest to lowest
summary_df = summary_df.sort_values("Total Profit", ascending=False)
summary_df
```
